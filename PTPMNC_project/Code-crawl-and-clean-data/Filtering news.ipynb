```python
from article_managers import Articles, ArticleManagers 
```

```python
excluded_articles_file_path = r'outputs/excluded_links.txt'
output_folder_path = r"outputs/cafef/"
processed_submaps_file_path = r'outputs/processed_submaps.txt'
sitemap_url = r'https://cafef.vn/sitemap.xml'
filtered_output_folder_path = r"outputs/cafef/filtered/"

am = ArticleManagers(excluded_articles_file_path = excluded_articles_file_path)
```

```python
processed_submaps = am.get_articles(processed_submaps_file_path)
processed_submaps
```

```python
submaps_path_to_process = []
for submap in processed_submaps:
    extracted_name, _, processed_name = am.get_file_name_from_submap(submap)
    processed_articles_file_path =  output_folder_path + processed_name
    filtered_articles_file_path = filtered_output_folder_path + extracted_name + '_filtered.txt'
    submaps_path_to_process.append([processed_articles_file_path, filtered_articles_file_path])

submaps_path_to_process
```

```python
def contains_text(text, word_list):
    for word in word_list:
        if word in text:
            return True
    return False
```

```python
wanted_categories = ['Doanh nghip',
                     'Th trng chng khon']

other_considerations = ['Bt ng sn',
                        'Ti chnh - ngn hng']

excluding = ['Doanh nghip gii thiu']


for submap_to_process in submaps_path_to_process:
    filtered_articles = []
    procesed_articles_path = submap_to_process[0]
    save_path = submap_to_process[1]
    articles = am.get_articles(procesed_articles_path)

    for article in articles:
        if contains_text(article.categories, wanted_categories) and not contains_text(article.categories, excluding):
            filtered_articles.append(article)

    am.save_articles(filtered_articles, save_path)
```
